## 3c
Make a dataframe for the year 2020 which contains the number of dentists per country and the number of inhabitants per country. <br>
(Hint: Use the Pandas function *merge*.)

Please name the dataframe dentists_population_2020. <br>
Please name the columns as follows:
* *Country Name*: Name of Country
* *Country Code*: Code of Country
* *Dentists*: Number of dentists in country in 2020
* *Population*: Population of Country in 2020

```python
population_2020 = population[["Country Name", "Country Code", "2020"]]
dentists_2020 = dentists[["Location", "SpatialDimValueCode", "Period", "FactValueNumeric"]]
dentists_2020 = dentists_2020[dentists_2020["Period"] == 2020]
```

```python
dentists_2020.head()
```

```python
population_2020.head()
```

```python
dentists_2020[dentists_2020["Location"] == "Switzerland"]
```

```python
dentists_population_2020 = pd.merge(dentists_2020, population_2020, left_on="SpatialDimValueCode", right_on="Country Code")
```

```python
dentists_population_2020 = dentists_population_2020[["Country Name", "Country Code", "2020", "FactValueNumeric"]]
dentists_population_2020.head()
```

```python
dentists_population_2020 = dentists_population_2020.rename(columns={"FactValueNumeric": "Dentists", "2020": "Population"})
dentists_population_2020.head()
```
